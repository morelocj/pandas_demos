```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from lxml import etree
```

Make Our Get Reuqest

```python
url= 'https://en.wikipedia.org/wiki/List_of_largest_banks'
response= requests.get(url)
response
```

Get content of our response variable

```python
html_data= response.content
html_data[:50]
```

Now we can Create a BeatifulSoup object to parse the html string

```python
soup= BeautifulSoup(html_data)
soup
```

Export into an html file

```python
with open('list_of_banks.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```
