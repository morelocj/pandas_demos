```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_proMo = df[['total_profit','month_number']]
```

```python
#df_proMo.set_index('month_number', inplace=True)
df_proMo.plot(kind = 'bar', legend=None, width=0.8, figsize = (7,5))
plt.title('Company profit per month',fontsize=14)
plt.xlabel('Month number',fontsize=12)
plt.ylabel('Total profit',fontsize=12)
plt.xlim(-1,12)
plt.xticks(np.arange(1,13,step=2))
plt.yticks(np.arange(100000,600000,step=100000))
plt.rc('grid', linestyle = '--')
plt.grid(True)
plt.show()
```

```python
df_proMo.plot(xlabel='Month number',ylabel='Total profit',title='Company profit per month', color='red', marker='o', mec = 'black',mfc='black',linewidth=3,linestyle='dashed')
plt.xticks(np.arange(1,13,step=1))
plt.yticks(np.arange(100000,600000,step=100000))
plt.legend(['Profit data of last year'],loc='lower right')
```

```python
df_ml = df.loc[:,['month_number', 'facecream', 'facewash', 'toothpaste', 'bathingsoap',
       'shampoo', 'moisturizer']]
df_ml.set_index('month_number',inplace=True)
df_ml.rename(columns={'facecream':'Face cream Sales Data', 'facewash':'Face Wash Sales Data', 'toothpaste':'ToothPaste Sales Data', 'bathingsoap':'bathingsoap Sales Data','shampoo':'Shampoo Sales Data', 'moisturizer':'Moisturizer Sales Data'},inplace=True)
df_ml
```

```python
ax = df_ml.plot(xlabel="Month Number", ylabel="Sales units in number",kind='line', figsize=(6,4),marker='o',linewidth=3)
ax.set_xticks(np.arange(1,13,step=1))
ax.set_yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
```

```python
df_ml.reset_index(inplace=True)
df_bsoap=df_ml[['month_number','bathingsoap Sales Data']]
df_bsoap.head()
```

```python
df_bsoap.plot(kind='scatter', s=100, x='month_number',y='bathingsoap Sales Data',figsize=(9,6))
plt.xlabel('Month Number',fontsize=15)
plt.ylabel('Number of units sold',fontsize=14)
plt.xticks(np.arange(1,13,step=1))
plt.title('bathingsoap Sales data',fontsize=18)
plt.rc('grid', linestyle = '--',linewidth='1')
plt.grid(True)
plt.legend(['bathing soap Sales data'],fontsize=15,loc='upper left')
plt.show()
```

```python
date=["25/12","26/12","27/12"]
temp=[8.5,10.5,6.8]
```

```python
df_temp = pd.DataFrame(date,temp)
df_temp.reset_index(inplace=True)
df_temp.rename(columns = {'index':'temp',0:'date'},inplace=True)
df_temp
```

```python
df_temp.plot(kind='line',x='date',y='temp',figsize = (9,6))
plt.xlabel('Date')
plt.ylabel('temperature')
plt.title('Date-wise Temperature')
plt.show()
```

```python
height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]
```

```python
df_size = pd.DataFrame(weight,height)
df_size.reset_index(inplace=True)
df_size.rename(columns = {'index':'height',0:'weight'},inplace=True)
df_size
```

```python
df_size.plot(kind='line',x='weight',y='height',figsize = (6,4), marker='o', ms=10, mec='g', linestyle='-.', linewidth=2,color='green',legend=None)
plt.xlabel('Weight in kg',fontsize=12)
plt.ylabel('Height in cm',fontsize=12)
plt.title('Average weight with respect to average height',fontsize=13)
plt.show()
```
